## Question 4

Which line creates a DataFrame containing only rows where `City` is `"New York"`?

A. `new_york_df = registrations["City"]`  
B. `new_york_df = registrations[registrations["City"] == "New York"]`  
C. `new_york_df = registrations["New York"]`  
D. `new_york_df = pd.read_csv("New York")`

````python
new_york_df = registrations[registrations["City"] == "New York"]
new_york_df
````

````python
print("Rows where City is New York:", len(new_york_df))
````

## Question 5

Which line updates `status` to `"Adult"` only for rows where `Age` is 18 or older?

A. `registrations["Age"] >= 18 = "Adult"`  
B. `registrations.loc[registrations["Age"] >= 18, "status"] = "Adult"`  
C. `registrations[registrations["status"] == "Adult"] = 18`  
D. `registrations.loc["status", "Age"] = "Adult"`  

Answer: B

````python
adult_demo = registrations.copy()
adult_demo.loc[adult_demo["Age"] >= 18, "status"] = "Adult"
adult_demo
````

````python
adult_demo["status"].value_counts()
````
