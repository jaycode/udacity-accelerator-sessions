# pandas Code Examples

Use this notebook with slide 8 of the Real World Data Wrangling session. By the end, learners can load a CSV, pull one column as a Series, filter rows, and update values with `.loc`.

## 1. Import pandas

The alias `pd` is the common convention, so most examples in the pandas documentation use it.

```python
# Import the pandas module to our code and reference it as pd
import pandas as pd
```

## 2. Load `data.csv` into a DataFrame

A DataFrame is pandas' table object. The file is in the same folder as this notebook, so the relative path is just `data.csv`.

```python
# Load a CSV file data.csv into a pandas DataFrame instance called df
df = pd.read_csv('data.csv')
df
```

## 3. Select one column as a Series

A Series is pandas' one-column object. The bracket style is the safer habit because it also works for column names with spaces.

```python
# Get a Series from that DataFrame object, which is the entire "Age" column.
ages = df['Age']
ages
```

```python
# Alternatively, you can also get it with the dot operator.
# It does not work when the column name has spaces, and does not look
# particularly nice when there are capital letters.
ages = df.Age
ages
```

## 4. Filter rows with a condition

Before running the next cell, predict how many rows have `Age < 11`.

```python
# Select rows where the value of the column "Age" is <11
children_df = df[df['Age'] < 11]
children_df
```

## 5. Update selected rows with `.loc`

Use `.loc[row_condition, column_name]` when you want to update a specific column for the rows that match a condition.

```python
# Update the "status" column of rows where the column "Age" is <11 to "Children".
# Notice the use of .loc.
df.loc[df['Age'] < 11, 'status'] = 'Children'
df
```

## Quick check

This final cell counts the status values so learners can confirm exactly which rows changed.

```python
df['status'].value_counts()
```
